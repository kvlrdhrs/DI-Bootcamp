```python
import pandas as pd
from plotnine import *
import numpy as np
```

1. Exercise 1 : Basic Scatter Plot With Plotnine

```python
x = np.random.randint(low = 100, high = 1000, size = 20)
y = np.random.randint(low = 100, high = 1000, size = 20)
df1 = pd.DataFrame({'x':x, 'y':y})
df1.head()
```

```python
scatter_plot = (
    ggplot(df1, aes(x='x', y='y')) + geom_point() + geom_smooth(method = 'lm') +
    labs(x='X-Value', y='Y-Value', title = 'Basic Scatter Plot')
)
print(scatter_plot)
```

Exercise 2 : Bar Plot Of Categorical Data With Plotnine

```python
df2 = pd.DataFrame({'category':np.random.choice(['A','B','C'], 30)})
```

```python
bar_plot = (
    ggplot(df2, aes(x='category', fill='category')) + 
    geom_bar() + 
    labs(x='X-Value', y='Y-Value', title='Category Frequency') + 
    geom_text(aes(label='stat(count)'), stat="count", nudge_y=0.125, va="bottom")
)

print(bar_plot)
```

 Exercise 3 : Line Plot With Multiple Lines With Plotnine

```python
x_val = np.linspace(0, 20, 1000)
y1_val = np.sin(x_val)
y2_val = np.cos(x_val)
df_3 = pd.DataFrame({'x': x_val, 'y1': y1_val, 'y2': y2_val})
```

```python
plot = (
    ggplot(df_3, aes(x='x')) +
    geom_line(aes(y='y1'), color='blue', size=1.5) +  # Plot y1 with blue color
    geom_line(aes(y='y2'), color='red', size=1.5) +   # Plot y2 with red color
    labs(x='X-Value', y='Y-Value', title='Comparison of Two Functions')
)


# Display the plot
print(plot)
```

Exercise 4 : Basic Line Chart With Plotly

```python
import plotly.graph_objects as go
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]
```

```python
fig = go.Figure(data=go.Scatter(x=months, y=sales))
fig.update_layout(title_text='Monthly Sales Data', title_x = 0.5)
fig.show()
```

Exercise 5 : Bar Chart With Customization With Plotly

```python
genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
```

```python
fig = go.Figure(data=go.Bar(x=genres, y=books_sold, text = books_sold))
fig.update_layout(title_text='Books Sold by Genre')
fig.show()
```

Exercise 6 : Scatter Plot With Size And Color Dimensions With Plotly

```python
import plotly.express as px
data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})
```

```python
fig = px.scatter(data, x='Age', y='Salary', size='Years of Experience', title='Relationship between Age, Salary, and Experience')

# Display the plot
fig.show()
```
