```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# Matplotlib Task: Generate a histogram for petal lengths
plt.figure(figsize=(8, 6))
plt.hist(df['petal_length'], bins=20, color='skyblue')
plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn Task: Produce a pair plot to show pairwise relationships in the dataset
sns.pairplot(df, hue='species', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```
