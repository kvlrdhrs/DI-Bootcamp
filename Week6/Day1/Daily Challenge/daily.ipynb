```python
import numpy as np
import pandas as pd
```

```python
# Data Preparation
a = np.random.uniform(-5,36, 120).reshape(10,12).astype(int)
```

```python
df = pd.DataFrame(a, columns=[f'month_{i}' for i in range(1,13)], index=[f'city_{i}' for i in range(1,11)])
df
```

```python
# Data Analysis

# annual avg temperature for each city 
df.mean(axis=1)
```

```python
# the city with the highest avg temparature

print(f"The city with highest avg temp.: {df.mean(axis=1).idxmax()}")
print(f"The city with lowest avg temp.: {df.mean(axis=1).idxmin()}")
```

```python
# Vizualization
import matplotlib.pyplot as plt
import seaborn as sns

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, cmap='coolwarm', annot=True, fmt=".1f", linewidths=0.5)
plt.title('Heatmap of Monthly Data for Cities')
plt.xlabel('Months')
plt.ylabel('Cities')
plt.show()
```

```python
# Line plot
plt.figure(figsize=(10, 6))
for city in df.index:
    plt.plot(df.columns, df.loc[city], marker='o', label=city)
plt.title('Monthly Data for Cities')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
