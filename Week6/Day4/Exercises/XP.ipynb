```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Exercise 1 : Linear Regression Analysis

house_sizes = [50, 70, 80, 100, 120]
house_prices =[150000, 200000, 210000, 250000, 280000]

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
fig, ax = plt.subplots(figsize = (6,6))
ax.scatter(house_sizes, house_prices, label='Data')

# Plotting the regression line
x_values = np.array([min(house_sizes), max(house_sizes)])
y_values = slope * x_values + intercept
plt.plot(x_values, y_values, color='red', label='Regression Line')

plt.xlabel('House Size (sq. meters)')
plt.ylabel('House Price ($)')
plt.title('House Size vs. House Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Predict the price of a house that is 90 square meters.
p1 = slope * 90 + intercept
print(f" The price of a 90 sm huse is: {int(p1)}")
```

```python
# Interpret the meaning of the slope in the context of housing prices.
# The slope in this example is 1828 (positive), it means that if house sizes increase house prices would be increase aswell. 
```

```python
group1 = [5, 6, 7, 6, 5]
group2 = [7, 8, 7, 9, 8]
group3 = [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(group1, group2, group3)
print("f-value: ", f_value)
print("p-value: ", p_value)
```

```python
# The groups show significant differences, and the null hypothesis that there are no differences between the groups is not valid.
# if p-value were greater than 0.05 it would be means that null hypothesis valid and we dont have differences between the groups in data.
```

```python
# Exercise 3 : Probability Distributions
n, p = 10, 0.5
binom_dist = stats.binom(n,p)
print("PMF at 5 successes:", binom_dist.pmf(5))

x_binom = np.arange(0, n+1)
pmf_binom = binom_dist.pmf(x_binom)
plt.figure(figsize=(6, 6))
plt.bar(x_binom, pmf_binom)
plt.title('Binomial Distribution\n$n=10$, $p=0.5$')
plt.xlabel('Number of Successes')
plt.ylabel('PMF')
plt.show()
```

```python
# Exercise 4 : Correlation Coefficients
import pandas as pd

data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
pearason = stats.pearsonr(data['age'], data['income'])
spearman = stats.spearmanr(data['age'], data['income'])
print(pearason, '\n', spearman)
```

```python
# Exercise 5 : Linear Relationships

# Performing linear regression
slope5, intercept5, r_value5, p_value5, std_err5 = stats.linregress(data['age'], data['income'])

fig, ax = plt.subplots(figsize = (6,6))
ax.scatter(data['age'], data['income'], label='Data')

# Plotting the regression line
x_values5 = np.array([min(data['age']), max(data['age'])])
y_values5 = slope5 * x_values5 + intercept5
plt.plot(x_values5, y_values5, color='red', label='Regression Line')

plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs. Income')
plt.legend()
plt.grid(True)
plt.show()
```
